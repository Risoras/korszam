```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
V = (30*2.556*1e-10)**3 #volume
corr = 3.1561*1e-23  # corrects for using m = kb = 2.556*10^-10m = 1 in the cpp code, I chose helium so m = 4
N = 1000
kb = 1.38 * 1e-23  #boltzmann
rho = (1000*4/(6*1e23))/V
```

```python
file = open("T.txt")

fileT=[]
for line in file:
    fileT.append(float(line.strip()))

file.close()
```

```python
T = np.zeros(len(fileT))
for i in range (len(fileT)):
    T[i] = fileT[i]
```

```python
Tcorr = T*corr    #correct temperature
p = N*kb*Tcorr/V  #pressure
```

```python
Z = np.zeros(len(T))
for i in range(len(T)):
    Z[i] = p[i]/(rho*8.314/4*Tcorr[i])
```

```python
t=np.linspace(0,10,10000)
```

```python
plt.plot(t,Tcorr)
plt.xlabel("time [ps]")
plt.ylabel("temperature [K]")
#plt.ylim(0,180)
plt.grid()
plt.savefig("temperature.png")
```

```python
plt.plot(t, 1e-6*p)
plt.xlabel("time [ps]")
plt.ylabel("pressure [MPa]")
#plt.ylim(0,0.25)
plt.grid()
plt.savefig("pressure.png")
```
