```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
V = (30*2.556*1e-10)**3 #volume
corr = 3.1561*1e-23  # corrects for using m = kb = 2.556*10^-10m = 1 in the cpp code
```

```python
file = open("T.txt")

fileT=[]
for line in file:
    fileT.append(float(line.strip()))

file.close()
```

```python
T = np.zeros(len(fileT))
for i in range (len(fileT)):
    T[i] = fileT[i]
p = T/V
```

```python
Z = np.zeros(len(T))
for i in range(len(T)):
    Z[i] = p[i]*V/(1000*T[i])
```

```python
Tcorr = T*corr
```

```python
p = T/V
pcorr = Tcorr/V
```

```python
t=np.linspace(0,10,10000)
```

```python
plt.plot(t,Tcorr)
plt.xlabel("time [ps]")
plt.ylabel("temperature [K]")
#plt.ylim(0,180)
plt.grid()
plt.savefig("temperature.png")
```

```python
plt.plot(t, pcorr)
plt.xlabel("time [ps]")
plt.ylabel("pressure [Pa]")
#plt.ylim(0,0.25)
plt.grid()
plt.savefig("pressure.png")
```
